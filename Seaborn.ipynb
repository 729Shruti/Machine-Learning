```python
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
cars=sns.load_dataset('mpg')
```

```python
cars.shape
```

```python
cars.dropna()
```

```python
cars.isnull().count()
```

```python
#Univariate kde  (single variable)
```

```python
sns.kdeplot(cars.horsepower)
```

```python
sns.kdeplot(cars.horsepower,shade =True)
```

```python
sns.kdeplot(cars.horsepower,bw=10) #bw=bandwidth
```

```python
sns.kdeplot(cars.horsepower,bw=1)
```

```python
sns.kdeplot(cars.horsepower,cumulative=True)
```

```python
#bivarte kde
```

```python
sns.kdeplot(x='horsepower',y='mpg', data=cars)
```

```python
sns.kdeplot(x='horsepower',y='mpg', data=cars,shade=True)
```

```python
sns.kdeplot(x='horsepower',y='mpg', data=cars,n_levels=20)
```

```python
sns.set_style('whitegrid')
sns.kdeplot(x='horsepower',y='mpg',data=cars,shade=True,shade_lowest=False,cbar=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
diamonds=sns.load_dataset('diamonds')
```

```python
diamonds
```

```python
diamond=diamonds[diamonds.cut.isin(['Premium','Good'])&
         diamonds.color.isin(['D','F','J'])
         ].sample(n=100,random_state=22)
```

```python
sns.scatterplot(x='carat',y='price',data=diamond)
```

```python
sns.scatterplot(x='x',y='y',data=diamond)
```

```python
sns.set_style('darkgrid')
```

```python
sns.scatterplot(x='price',y='carat',data=diamond,hue='cut')
```

```python
sns.scatterplot(x='price',y='carat',hue='cut',palette=['purple','#55CCCC'],data=diamond)
```

```python
sns.scatterplot(x='price',y='carat',hue='depth',data=diamond)
```

```python
sns.scatterplot(x='price',y='carat',
                style='cut',
                s=50,data=diamond)
```

```python
sns.scatterplot(x='price',y='carat',
                size='cut',
                data=diamond)
```

```python
sns.scatterplot(x='price',y='carat',
                size='cut',sizes=[150,100],
                data=diamond)
```

```python
sns.scatterplot(x='price',y='carat',
                size='depth',
                data=diamond)
```

```python
sns.scatterplot(x='price',y='carat',
                hue='cut',
                style ='cut',
                data=diamond)
```

```python
sns.scatterplot(x='price',y='carat',
                hue='cut',
                style ='color',
                s=100,
                data=diamond)
```

```python
sns.scatterplot(x='price',y='carat',
                s=100,
                alpha=0.6 ,# vale lies between 0 to 1 t(transparency)
                
                data=diamond)
```

```python
sns.scatterplot(x='price',y='carat',
                s=200,
                marker='*',
                data=diamond)
```

```python
sns.scatterplot(x='price',y='carat',
                s=200,
                marker='*',
                edgecolor='black',
                data=diamond)
```
