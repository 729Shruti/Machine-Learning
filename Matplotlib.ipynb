```python
fig=plt.figure()
fig.suptitle('My Subplots')
fig.add_subplot(221)
plt.plot([np.log(n) for n in range (1,10)])
fig.add_subplot(222,facecolor='y')
fig.add_subplot(223)
fig.add_subplot(224)
plt.show()
```

```python
fig,plots=plt.subplots(2,sharex=True) #creating a figure with 2 subplots sharing same x axis
fig.suptitle('Sharing X axis')
x=range(0,200,5) # start to end and step of 5
y=[n**0.8 for n in x]
plots[0].plot(x,y,color='r') #first subplot
plots[1].scatter(x,y) #second subplot
plt.show()
```
